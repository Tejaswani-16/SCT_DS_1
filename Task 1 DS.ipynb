```python
import pandas as pd
import matplotlib.pyplot as plt
file="C:/Users/HP/OneDrive/Desktop/python/Copy of Sample - Superstore.xlsx"
df=pd.ExcelFile(file).parse(0)
print(df)
```

# Histogram

```python
plt.figure(figsize=(8,6))
plt.hist(df["Sales"], bins=20, edgecolor="black")
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

# Bar Chart

```python
c=df["Category"].value_counts()
plt.figure(figsize=(8,6))
c.plot(kind="bar", color=["lavender", "pink", "skyblue"], edgecolor="black")
plt.title("Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```
